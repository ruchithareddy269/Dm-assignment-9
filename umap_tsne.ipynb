```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import umap
import umap.plot
from sklearn.cluster import KMeans
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

# Imports
# from sklearn.decomposition import PCA
# from sklearn.manifold import TSNE
# from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
# import plotly.express as px


pd.plotting.register_matplotlib_converters()
%matplotlib inline
sns.set(style='darkgrid', font_scale=1.4)
```

```python
dataset_file_path = '/content/drive/MyDrive/wine-clustering.csv'
# Load the dataset
data = pd.read_csv(dataset_file_path)
```

### EDA

```python
# Displaying descriptive statistics
print(data.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizing the distribution of the 'Alcohol' feature using a box plot
plt.figure(figsize=(8, 4))
sns.boxplot(x=data['Alcohol'], color='green')
plt.title('Box Plot of Alcohol Distribution')
plt.xlabel('Alcohol')
plt.ylabel('Value')
plt.show()
```

UMAP Dimensionality Reduction

```python
# Scaling the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

Applying UMAP

```python
# Applying UMAP
umap_model = umap.UMAP(n_neighbors=5, min_dist=0.3, n_components=2, random_state=42)
umap_results = umap_model.fit_transform(scaled_data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizing the UMAP results with a density heatmap
plt.figure(figsize=(8, 5))
sns.kdeplot(x=umap_results[:, 0], y=umap_results[:, 1], cmap="Reds", shade=True, alpha=0.5)
plt.title('UMAP Dimensionality Reduction - Density Heatmap')
plt.xlabel('UMAP1')
plt.ylabel('UMAP2')
plt.show()
```

## Clustering

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(umap_results)
```

```python
# Visualizing the clustering results
plt.figure(figsize=(8, 5))
plt.scatter(umap_results[:, 0], umap_results[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.5)
plt.title('KMeans Clustering on UMAP results')
plt.xlabel('UMAP1')
plt.ylabel('UMAP2')
plt.show()
```

### Classification using Logistic Regression

```python
# Assuming the KMeans labels as target variable for classification
X_train, X_test, y_train, y_test = train_test_split(umap_results, kmeans.labels_, test_size=0.3, random_state=42)
```

Training Logistic Regression Model

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

Predictions

```python
# Making predictions
y_pred = lr.predict(X_test)
```

Model Evaluation

```python
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

### Summary of Analysis Conducted on Wine Dataset

This notebook provided a comprehensive analysis of a wine dataset through various data science techniques, leading to the following insights:

Exploratory Data Analysis (EDA):

The initial exploration of the dataset shed light on key features, especially focusing on aspects like the alcohol content. This phase helped in understanding the basic structure and characteristics of the dataset.
Dimensionality Reduction via UMAP:

To tackle the challenge of high-dimensional data, UMAP (Uniform Manifold Approximation and Projection) was employed. This method successfully condensed the data into a two-dimensional representation, revealing patterns and structures that were not apparent in the higher-dimensional space.
Clustering Analysis with KMeans:

Following the application of UMAP, the dataset was subjected to KMeans clustering. This approach effectively grouped the wines into distinct categories, indicating the presence of intrinsic similarities among certain types of wines.
Predictive Modeling with Logistic Regression:

Utilizing the clusters identified by KMeans as labels, a Logistic Regression model was developed. The performance of this model, evaluated through accuracy and other classification metrics, demonstrated its capability to reasonably predict the identified wine clusters.
